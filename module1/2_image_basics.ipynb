```python
import cv2
import argparse
import sys
sys.path.append("../")
import cv_imshow
```

```python
args = {
    "image": "../images/beach.png"
}
```

```python
#display image
image = cv2.imread(args["image"])
(h,w) = image.shape[:2]
cv_imshow.display_image("Original", image)
```

```python
#access pixel
(b, g, r) = image[0, 0]
print("Pixel at (0,0) - Red: {r}, Green: {g}, Blue: {b}".format(r=r, g=g, b=b))

#slicing image
(cX, cY) = (w//2, h//2)
tl = image[0:cY, 0:cX]
cv_imshow.display_image("Top-left corner", tl)
```

```python
tr = image[0:cY, cX:w]
br = image[cY:h, cX:w]
bl = image[cY:h, 0:cX]
cv_imshow.display_image("Top-Right Corner", tr)
cv_imshow.display_image("Bottom-Right Corner", br)
cv_imshow.display_image("Bottom-Left Corner", bl)
```
