# FAST
- Most simple and intuitive keypoint detector to understand, also one of fastest.
- Used to detect corners in images and is applicable to real-time applications or resource constrained devices.

## How does it work
- For a pixel to be considered a corner, there must be at least n contigous pixels along its circular perimeter with radius r that are all brighter or darker than center pixel by a threshold t.
<img src="../../images/embedded_images/fast_example.jpg" alt="Drawing" style="width: 300px;"/>
- In the above example image, we have center pixel value as 32. Let threshold be t=16, and n=12 contiguous pixels along boundary of circle that are all either brighter than $p+t$ or darker than $p-t$ for the center to be considered keypoint.
- As we can see, there are 14 contiguous pixels in the boundary that meets the stated condition, therefore the center pixel is a kepoint.

```python
import numpy as np
import cv2
import imutils
from cv_imshow import create_subplot, display_image
```

```python
args = {
    "image":"../../images/fast_book_cover.png",
    "pen":"../../images/keypoint_detect/pen.jpg"
}
```

```python
#load and convert to grayscale
image = cv2.imread(args["image"])
orig = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#detecting fast keypoints
detector = cv2.FastFeatureDetector_create()
kps = detector.detect(gray, None)

print("# of keypoints: {}".format(len(kps)))
```

```python
#loop over keypoints and draw them
for kp in kps:
    r = int(0.5*kp.size)
    (x,y) = np.int0(kp.pt)
    cv2.circle(image, (x,y), r, (0,255,0),2)
```

```python
display_image("Image",np.hstack([orig, image]))
```

```python
#load and convert to grayscale
image = cv2.imread(args["pen"])
orig = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#detecting fast keypoints
detector = cv2.FastFeatureDetector_create()
kps = detector.detect(gray, None)

print("# of keypoints: {}".format(len(kps)))

#loop over keypoints and draw them
for kp in kps:
    r = int(0.5*kp.size)
    (x,y) = np.int0(kp.pt)
    cv2.circle(image, (x,y), r, (0,255,0),2)
    
display_image("Image",np.hstack([orig, image]))
```
