```python
import import_ipynb
from object_detection.helpers import pyramid
import cv2
import sys
sys.path.append("../../../")
from cv_imshow import display_image, create_subplot
from matplotlib import pyplot as plt
```

```python
args = {
    "image":"../../../images/florida_trip.png",
    "scale":1.5
}
```

```python
fig = plt.figure(figsize=(10,10))
rows = 3
columns = 3

image = cv2.imread(args["image"])

for (i,layer) in enumerate(pyramid(image, scale=args["scale"])):
    create_subplot(fig, rows, columns, i+1, "Layer {}".format(i+1), layer)
```
